# 比特幣address的產生方式
## 要安裝bitcoin套件
參考網址：https://pypi.org/project/bitcoin/  
指令：pip install bitcoin

```python
from bitcoin import *
```

```python
# 隨機產生一組私鑰(2^256)
my_private_key = random_key()
print(my_private_key)
```

```python
# 利用私鑰產生公鑰
my_public_key = privtopub(my_private_key)
print(my_public_key)
```

```python
# 利用公鑰產生比特幣address
my_address = pubtoaddr(my_public_key)
print(my_address)
```
